```python
y_train
```

```python
X_test = test.loc[:, test.columns != 'contest-tmp2m-14d__tmp2m']
y_test = test.loc[:, test.columns == 'contest-tmp2m-14d__tmp2m']
```

```python
X_train = X_train.loc[:, X_train.columns != 'startdate']
X_test = X_test.loc[:, X_test.columns != 'startdate']
X_train = X_train.loc[:, X_train.columns != 'climateregions__climateregion']
X_test = X_test.loc[:, X_test.columns != 'climateregions__climateregion']
```
